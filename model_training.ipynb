## Importación de librerias

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importacion de datos

```python
dataset = pd.read_csv('Transform_Hospitalization.csv')
```

### Elimino las columnas de 'NO' --FIEBRE_NO, ITU_NO, TIPO DE CULTIVO_NO , AGENTE AISLADO_NO , PATRON DE RESISTENCIA_NO--

```python
# Selecciono las columnas para entrenar 'X' y la columna objetivo 'Y'
# Importante transformar a arrays para no tener qu eescalar a la hora de usar el modelo


X = dataset.iloc[:, :-1].values# menos la ultima columna
y = dataset.iloc[:, -11].values# toma la ultima columna

```

## División de datos en entrenamiento y prueba

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train[:11])
```

```python
print(y_train[:11])
```

## Escalado de funciones

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()# crea el objeto de escalado 
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train[:11])
```

```python
# sklearn.linear_model.LogisticRegression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state = 0) # creamos el objeto, y una semilla de randon cero 
model.fit(X_train, y_train)
```

## Matriz de confucion
